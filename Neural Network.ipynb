```python
#Normalising the data
from sklearn.preprocessing import MinMaxScaler
MM=MinMaxScaler()
for column in df.columns:
    if df[column].dtype == 'int':
        df[column]=MM.fit_transform(df[column])
df
    
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
for column in df.columns:
    if df[column].dtype == 'object':
        df[column]=LE.fit_transform(df[column])
df
```

```python
import matplotlib.pyplot as plt
plt.figure()
df.boxplot(figsize=(50,20))
plt.show()
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1

Lower_limit=Q1-1.5*IQR
Upper_limit=Q3+1.5*IQR

outliers=((df < Lower_limit) | (df > Upper_limit))
df_clean=df[~outliers.any(axis=1)]
df_clean
```

```python
X=df.drop(columns="letter")
Y=df["letter"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=45)

# Ensure data types and reshaping
import numpy as np
X_train = np.asarray(X_train).astype(np.float32)
Y_train = np.asarray(Y_train).astype(np.float32).reshape(-1,1)
X_test = np.asarray(X_test).astype(np.float32)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dropout 



# Constructing the ANN model
model = Sequential()
model.add(Dense(units=64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=8, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=4, activation='relu'))# Use 'softmax' for multi-class classification
model.add(Dropout(0.5))
model.add(Dense(units=2, activation='relu')) 
model.add(Dropout(0.5))
model.add(Dense(units=1, activation='softmax')) 
# Compile the model
optimizer=Adam(learning_rate=0.01)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model on the training set
model.fit(X_train, Y_train, epochs=150, batch_size=32, validation_split=0.2)

# Make predictions on the test set
Y_pred = model.predict(X_test)

print("Predictions on test set:", Y_pred)
```
