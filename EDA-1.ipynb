# Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Histogram is used to determine the shape of the data i.e,to find the skewness of the data
for i in df.select_dtypes(include='number').columns:
    sns.histplot(data=df,x=i)
    plt.show()
```

```python
#Boxplot is used to identify outliers in the data
```

```python
for i in df.select_dtypes(include='number').columns:
    sns.boxplot(data=df,x=i)
    plt.show()
```

```python
#Scatter plot is used to check the relationship between the two continous variables
```

```python
for i in ['AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV',
       'MLTV', 'Width', 'Tendency', 'NSP']:
    sns.scatterplot(data=df,x=i,y='LB')
    plt.show()
```

```python
#finding the correlation between the variables
df.select_dtypes(include='number').corr()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.select_dtypes(include='number').corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
